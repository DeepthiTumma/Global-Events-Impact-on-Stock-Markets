# Time Series Analysis

```python
print(df["Close_Price"].min())
print(df["Close_Price"].max())
```

```python
monthly_price = df.groupby(["year","Month"])["Close_Price"].mean().reset_index()
```

```python
year_month = (monthly_price["year"].astype(str) + "-" + monthly_price["Month"].astype(str))
```

```python
monthly_price["Year_month"] = year_month 
```

```python
monthly_price.columns
```

```python
import matplotlib.pyplot as plt

x = monthly_price["Year_month"]
y = monthly_price["Close_Price"]

plt.figure(figsize=(15,5))

plt.plot(x,y)

plt.xlabel("Year_month")
plt.ylabel("Close_Price")
plt.title("Time Series Analysis")

plt.xticks(ticks=range(0, len(x), 12), rotation=90)
plt.show()
```
